```python
def load_images_in_batches(input_dir, categories, batch_size):
    data = []
    labels = []
    count = 0
    
    for category_idx, category in enumerate(categories):
        category_path = os.path.join(input_dir, category)
        for file in os.listdir(category_path):
            img_path = os.path.join(category_path, file)
            
            # Ensure that only image files are processed
            if file.endswith('.jpg') or file.endswith('.png'):
                try:
                    # Read and resize the image
                    img = imread(img_path)
                    img = resize(img, (15, 15))  # Resize to 15x15 pixels

                    # Check if image has the correct shape
                    if img.shape == (15, 15, 3):  # Ensure it's a valid 3-channel RGB image
                        # Flatten the image and store it in data
                        data.append(img.flatten())
                        labels.append(category_idx)
                        count += 1
                    
                    # Yield a batch when the batch size is reached
                    if count == batch_size:
                        yield (np.asarray(data), np.asarray(labels))  # Convert to NumPy arrays
                        data, labels = [], []  # Reset for the next batch
                        count = 0
                
                except Exception as e:
                    # Print the error and skip the file
                    print(f"Error processing {img_path}: {e}")

    # Yield any remaining data that didn't fill up a batch
    if data and labels:
        yield (np.asarray(data), np.asarray(labels))

# Load the entire dataset into memory
def load_full_dataset(input_dir, categories, batch_size):
    data = []
    labels = []
    
    # Iterate through batches
    for data_batch, label_batch in load_images_in_batches(input_dir, categories, batch_size):
        data.extend(data_batch)
        labels.extend(label_batch)
        print(f"Processed batch with {data_batch.shape[0]} images and labels.")
    
    return np.asarray(data), np.asarray(labels)
# Example usage of batch processing function
input_dir = r'C:\Users\HP\Documents\kagglecatsanddogs_5340\PetImages'
categories = ['Cat', 'Dog']
batch_size = 128  # Define your batch size here
   

# Load the entire dataset into memory
data, labels = load_full_dataset(input_dir, categories, batch_size)
```

# Splitting data into Training and Testing examples

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data,labels, test_size= 0.2, random_state = 42)
```

# Building the SVM model for classification

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Step 2: Standardize the features 
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Step 3: Create the SVM classifier
svm_clf = SVC(kernel='linear', C=1.0) 
```

```python
# Step 4: Train the SVM classifier
svm_clf.fit(X_train_scaled, Y_train)
```

```python
# Step 5: Evaluate the model on the test set
y_pred = svm_clf.predict(X_test_scaled)
y_pred
```

```python
# Step 6: Output accuracy and classification report
accuracy = accuracy_score(Y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# Print detailed classification report
print(classification_report(y_test, y_pred, target_names=categories))
```
